```python
#Resample files to 22khz and 16 bits
# List all files ending with .wav
import librosa
import soundfile as sf
import numpy as np

peter_dir = os.path.join(base_dir, "Data/peter_data")
peter_new_dir = os.path.join(base_dir, "Data/peter_low_data")

wav_files = [f for f in os.listdir(peter_dir) if f.lower().endswith('.wav')]
# Sort files to ensure consistent ordering
wav_files.sort()


# resample each file to a sequential number
for idx, filename in enumerate(wav_files, start=1):
    # Load and resample to 22050 Hz
    src = os.path.join(peter_dir, filename)
    dest = os.path.join(peter_new_dir, filename)
    y, sr = librosa.load(src, sr=16000)
    # Convert to 16-bit PCM
    y_16bit = (y * 32767).astype(np.int16)
    # Save the output
    sf.write(dest, y_16bit, 16000, subtype='PCM_16')
    print(f"Resampling '{src}'")

print("finished")
```

```python
import os
import csv
import whisper

# Define the path to your merged "peter" folder
peter_folder = os.path.join(base_dir, "peter")

# Load the Whisper model ("turbo" variant)
print("Loading Whisper turbo model...")
model = whisper.load_model("turbo")

# Prepare a list to hold transcription results
transcriptions = []

# List the .wav files in the "peter" folder
wav_files = sorted([f for f in os.listdir(peter_folder) if f.lower().endswith('.wav')])

print(f"Found {len(wav_files)} .wav files. Beginning transcription...")

# Iterate over each .wav file, transcribe it, and store the result
for wav_file in wav_files:
    file_path = os.path.join(peter_folder, wav_file)
    print(f"Transcribing {wav_file}...")
    result = model.transcribe(file_path)
    transcription_text = result["text"].strip()
    # Append a dictionary with the track_id and its transcription
    transcriptions.append({'track_id': wav_file, 'transcription': transcription_text})

# Specify the output CSV file path in the base directory
csv_filename = "transcriptions_peter.csv"
csv_path = os.path.join(base_dir, csv_filename)

# Write the transcriptions to the CSV file
with open(csv_path, "w", newline="", encoding="utf-8") as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=['track_id', 'transcription'])
    writer.writeheader()
    for entry in transcriptions:
        writer.writerow(entry)

print(f"Transcription complete. CSV file created at: {csv_path}")
```
