```python
import whisper

# Load the Whisper model (using the "turbo" model here)
model = whisper.load_model("turbo")
```

```python
#Audio Recording Function

from IPython.display import Audio, display, Javascript
from google.colab.output import eval_js
import base64

def record(sec=5):
  js_code = """
    async function recordAudio() {
      const stream = await navigator.mediaDevices.getUserMedia({ audio: true });
      const mediaRecorder = new MediaRecorder(stream);
      let chunks = [];
      mediaRecorder.ondataavailable = e => chunks.push(e.data);
      mediaRecorder.start();
      await new Promise(resolve => setTimeout(resolve, %d * 1000));
      mediaRecorder.stop();
      await new Promise(resolve => mediaRecorder.onstop = resolve);
      let blob = new Blob(chunks, { 'type' : 'audio/wav' });
      let reader = new FileReader();
      reader.readAsDataURL(blob);
      await new Promise(resolve => reader.onload = resolve);
      return reader.result;
    }
    recordAudio();
  """ % sec
  print("Recording for {} seconds...".format(sec))
  audio_data = eval_js(js_code)
  header, encoded = audio_data.split(',', 1)
  data = base64.b64decode(encoded)
  return data
```

```python
#Record Audio
# Record 5 seconds of audio and save it as 'recorded.wav'
audio_bytes = record(5)
with open('recorded.wav', 'wb') as f:
  f.write(audio_bytes)
print("Recording complete!")
```

```python
#Playback

# Display the recorded audio for playback
display(Audio('recorded.wav', autoplay=True))
```

```python
#Transcribe

# Transcribe the recorded audio file and print the result
result = model.transcribe("recorded.wav")
print("\nTranscription:")
print(result["text"])
```
